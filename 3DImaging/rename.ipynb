```python
def rename_files_in_folder_with_annotation(folder_path, new_name_prefix):

    files = os.listdir(folder_path)
    counter = 0
   

    for file in files:

        formatted_counter = str(counter).zfill(len(str(999)))

        new_file_name = os.path.join(folder_path, f"{new_name_prefix}_{formatted_counter}.nii.gz")
        old_file_name = os.path.join(folder_path, file)

        print(old_file_name)
        print(new_file_name)

        os.rename(old_file_name, new_file_name)

        counter += 1
        
     

    return files


base_folder = '.'

#rename_files_in_folder_with_annotation("./imagesTr", "aneurysm")
#rename_files_in_folder_with_annotation("./imagesTs", "aneurysm")

rename_files_in_folder_with_annotation("./labelsTr", "aneurysm")
rename_files_in_folder_with_annotation("./labelsTs", "aneurysm")

#rename_files_in_folder_with_annotation("./val", "artery")
#rename_files_in_folder_with_annotation("./valannot", "artery_gt")
```
